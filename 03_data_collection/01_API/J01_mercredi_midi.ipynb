## 6 - Given the previous steps, create a dictionary containing the historical OHLC prices for the ten first coins over the running Year-To-Date period

* Dictionary
* The ten first coins 
* Over the running Year-To-Date period

```python
def getYtDData(coinCompany):
  url   = "https://api.coingecko.com/api/v3/"
  route = f"coins/{coinCompany}/ohlc"
  params = {
    "vs_currency" :"usd", 
    "days"       :"365", 
    "precision"    :"2", 
  }
  headers = {
    "accept" : "application/json",
  }

  my_resp = requests.get(url+route, headers = headers, params=params)
  print('status code:', my_resp)

  # on récupère une liste de listes (matrice)
  my_list = my_resp.json()
  return my_list

def convertDate(my_list):
  for line in my_list:
    timestamp = datetime.datetime.fromtimestamp(line[0]/1000)
    line[0] = timestamp.strftime('%Y-%m-%d')
  return my_list  
```

```python
# Solution 1 avec le DataFrame des 10 1er sociétés
# Ca marche pour 3 mais ça bloque au dessus
# Erreur 429 : Too Many Requests
df_tmp = df10[:3]

top10YtD_dico ={}
for company in df_tmp["id"]:
  time.sleep(1)
  print(company)
  # get the YtD data
  YtD_list = getYtDData(company)
  # convert the date
  YtD_list = convertDate(YtD_list)

  # in the top10YtD_dico
  # for each section company
  # create a dict where key is the date and value the list of 4 values OLHS 
  top10YtD_dico[company] = {}
  for line in YtD_list:
    date = line[0]
    line.remove(line[0])
    top10YtD_dico[company].update({date: line})

top10YtD_dico  
  
```
