### Using plotly.express methods' arguments to refine the figures

As in seaborn, the methods listed above have a bunch of optional arguments that allow to add some information to the figures. Depending on the method you are using, the available options can be different. Please refer to the documentation to know which options are available for a given method. Let's go through some examples with the `px.scatter` method :

#### Color

You can change the marker color depending on a given categorical column of your dataset :

```python
fig = px.scatter(df, x = "sepal_width", y = "sepal_length", color = "species")
fig.show()
```

```python
# import seaborn as sns
# import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8,5))
sns.scatterplot (df, x='sepal_width', y='sepal_length', hue='species', alpha=.6, legend=True)
# ax.set_title("Sepal Length")
# ax.set_xlabel("Sepal Width")
```

#### Changing the colorscale
You can change the color scale, by using either a discrete color sequence or a continuous colorscale. There are plenty of built-in colorscales available in Plotly.

You can refer to these documentation pages:
* [Discrete colors in Plotly](https://plotly.com/python/discrete-color/)
* [Continuous colorscales](https://plotly.com/python/colorscales/)
* [Built-in continuous colorscales](https://plotly.com/python/builtin-colorscales/)

```python
# Discrete sequence
fig = px.scatter(df, x = "sepal_width", y = "sepal_length", color = "species", color_discrete_sequence = px.colors.qualitative.Pastel)
fig.show()
```

```python
# import seaborn as sns
# import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8,5))
sns.scatterplot (df, x='sepal_width', y='sepal_length', palette="Set1", hue='species', alpha=.6, legend=True)
ax.set_title("Sepal Length")
ax.set_xlabel("Sepal Width")
```

```python
# Continuous sequence - column passed in "color" must be numeric
fig = px.scatter(df, x = "sepal_width", y = "sepal_length", color = "species_id", color_continuous_scale = px.colors.sequential.Plasma)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
sns.scatterplot (df, x='sepal_width', y='sepal_length', palette="rocket", hue='species_id', alpha=.6, legend=True)
ax.set_title("Sepal Length")
ax.set_xlabel("Sepal Width")
```

#### Trendline
You can add a trendline as in the following :

```python
fig = px.scatter(df, x = "sepal_width", y = "sepal_length", color = "species", trendline = "ols")
fig.show()
```

```python
# import seaborn as sns
# import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8,5))
sns.regplot(df, x='sepal_width', y='sepal_length')
# sns.scatterplot (df, x='sepal_width', y='sepal_length', palette="Set1", hue='species', alpha=.6, legend=True)
ax.set_title("Sepal Length")
ax.set_xlabel("Sepal Width")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# fig, ax = plt.subplots(figsize=(8,5))
sns.lmplot(df, x='sepal_width', y='sepal_length', hue="species")
# sns.scatterplot (df, x='sepal_width', y='sepal_length', palette="Set1", hue='species', alpha=.6, legend=True)
ax.set_title("Sepal Length")
ax.set_xlabel("Sepal Width")
```

#### 3D visualisation
There's also a `scatter_3d` method that allows you to add one more dimension to your visualisation :

```python
fig = px.scatter_3d(df, x = "sepal_width", y = "sepal_length", z = "petal_length", color = "species")
fig.show()
```

```python
fig = plt.figure(figsize=(10,10))
plot_axes = plt.axes(projection = '3d')

x = df["sepal_width"]
y = df["sepal_length"]
z = df["petal_length"]

plot_axes.scatter3D(x, y, z)

plot_axes.set_xlabel('x')
plot_axes.set_ylabel('y')
plot_axes.set_zlabel('zz')

plt.show()
```

#### Marginals
Let's go back to the 2D scatter plot. You can add some summarized information by using the `marginal_x` and `marginal_y` arguments :

```python
fig = px.scatter(df, x = "sepal_width", y = "sepal_length", color = "species", trendline = "ols", marginal_x = "violin", marginal_y = "box")
fig.show()
```

#### Facets
You can use `facet_row` and `facet_col` to create some facets, each containing the data for a given category. If you're using it together with the `marginal_x` and `marginal_y` options, the marginals won't be displayed :

```python
# Create one subgraph for each category and set the dislay order
fig = px.scatter(df, x = "sepal_width", y = "sepal_length", color = "species", trendline = "ols", marginal_x = "violin", marginal_y = "box", facet_row = "species")
fig.show()
```

```python
# Create one subgraph for each category and set the dislay order
fig = px.scatter(df, x = "sepal_width", y = "sepal_length", color = "species", trendline = "ols", marginal_x = "violin", marginal_y = "box",
                 facet_row = "species", facet_col = "species_id")
fig.show()
```

#### Categories order
Finally, you can change the order of the categories that you are using with the facets thanks to `category_orders` :

```python
category_orders_dict = {
    "species" : ["setosa", "virginica", "versicolor"],
    "species_id" : [3, 2, 1]
}

fig = px.scatter(df, x = "sepal_width", y = "sepal_length", color = "species", trendline = "ols", marginal_x = "violin",
                 marginal_y = "box", facet_row = "species", facet_col = "species_id", category_orders = category_orders_dict)
fig.show()
```

## The different types of charts

Now that you've become familiar with Plotly Express, let's explore the main different types of charts and their applications.

### Scatter plots
All the examples above were illustrating a scatter plot.

Scatter plots are best for displaying values for two variables from a dataset. They are great for **showing the overall relationship between the two variables** in a large amount of data.

The data is displayed as a collection of points; the value of one variable determines the position on the horizontal axis, while the value of the other variable determines the position on the vertical axis.

### Line plots
Line charts are best used when trying to visualize **numeric data over time**. Line charts are set against a common scale and are ideal for showing trends in data.

```python

# ! Ne pas oublier le reset_index() 
df = (px.data.gapminder()).reset_index()
df.head()
type(type(df))
df.shape
print(df.index)
display(df)
```

```python
# Create df for France
# df_france = df.loc[df['country']== "France", :]
df_france = df.loc[df['country']== "France"] # Comme on a fait un reset_index() pas besoin de preciser les colonnes
```

```python
fig = px.line(df_france, x = "year", y = "lifeExp")
fig.show()
```

```python
# Create df for Oceania
df_oceania = df.loc[df['continent']== "Oceania", :]
```

```python
# Remember there exists lots of extra arguments you can use to refine your plot
fig = px.line(df_oceania, x = "year", y = "lifeExp", color="country")
fig.show()
```

## Histograms
Histograms are appropriate to show the **distribution of a numeric variable**.
An important parameter is the number of bins, which will define the "level of details":
* For small datasets, the number of bins must be small
* For bigger datasets, the number of bins can be higher

```python
df = px.data.tips()
df.head()
```

```python
fig = px.histogram(df, x = "tip")
fig.show()
```

```python
# Number of bins to low
fig = px.histogram(df, x = "tip", nbins = 5)
fig.show()
```

```python
# Number of bins to high
fig = px.histogram(df, x = "tip", nbins = 500)
fig.show()
```

## Bar charts
Bar charts are best used when showing **comparisons between categories**. Typically, the bars are proportional to the values they represent and can be plotted either horizontally or vertically. One axis of the chart shows the specific categories being compared, and the other axis represents numeric values. Bar charts are ideal when you’re working with limited space.

```python
# Average bill grouped by day
df_day = df.groupby('day')['total_bill'].mean().reset_index()
df_day.head()
```

```python
fig = px.bar(df_day, x = "day", y = "total_bill")
fig.show()
```

```python

# ! ordonner les jours correctement
cat_orders = {
  'day' : ["Thur", "Fri", "Sat", "Sun"], 
}

fig = px.bar(df_day, x = "day", y = "total_bill", category_orders=cat_orders)
fig.show()
```

```python
# Average bill grouped by day and time
df_day_time = df.groupby(['day','time'])['total_bill'].mean().reset_index()
df_day_time.head()
```

```python
# px.bar will stack bars by default
fig = px.bar(df_day_time, x = "day", y = "total_bill", color = "time")
fig.show()
```

```python
# You can use the barmode argument to change this behaviour
fig = px.bar(df_day_time, x = "day", y = "total_bill", color = "time", barmode = "group")
fig.show()
```

```python
# Change the orientation of the bars - you must swap x and y
fig = px.bar(df_day_time, x = "total_bill", y = "day", color = "time", barmode = "group", orientation = "h")
fig.show()
```

## Boxplots
Whisker Charts or Box Plots are very useful to **compare distributions of numeric variables across different categories**.

```python


# ! ATTENTION : Sur la première boxplot je retrouve pas le 1.5*IQR pour la moustache haute (pour celle du bas il prend le min de la série, c'est OK)
fig = px.box(df, x = 'day', y = 'total_bill', color = 'time')
fig.show()
```

## Pie charts
Pie charts are best used to compare parts to the whole. Pie charts make it easy for an audience to understand the relative importance of values, but when there are more than five sections, it can become difficult to compare the results. The difference between the sections can become too narrow to effectively interpret.

```python
# Compute total revenue cumulated by day
# ! Attention au reset_index()
df_revenue = df.groupby('day')['total_bill'].sum().reset_index()
df_revenue.head()
```

```python
fig = px.pie(df_revenue, names = "day", values = "total_bill")
fig.show()
```

## Bubble charts
Bubble Charts are used to show **three dimensions of data—comparing entities** in terms of their relative values, positions, and sizes. Bubble charts are similar to scatter plots, where the data points are replaced with bubbles.

```python
df.head()
```

```python
# To create a bubble chart, just use the argument size of the px.scatter function
fig = px.scatter(df, x = "total_bill", y = "tip", size = "size", color = "smoker")
fig.show()
```

## Geographical data
If the dataset contains information about latitude and longitude, you can use px.scatter_mapbox to display the data on a map!

```python
df_car = px.data.carshare()
df_car.head()
```

```python
fig = px.scatter_mapbox(df_car, lat="centroid_lat", lon="centroid_lon", color="peak_hour", size="car_hours", zoom=10, mapbox_style="carto-positron")
fig.show()
```

### Making awesome animations
Last but not least : it's possible to create animated graphs with Plotly ! You just need to set the animation_frame and (optionnaly) animation_group arguments:

```python
df = px.data.gapminder()
df.head()
```

```python
fig = px.scatter(df, x = "gdpPercap", y = "lifeExp", animation_frame = "year", animation_group = "country", log_x = True)
fig.show()
```

```python
# Faut penser à ajuster les gammes max des valeurs en x et y pour toutes les années
# Sinon ça sort
# Voir range_x = [100,100000], range_y = [25,90]
# Il faut que la DataFrame ait une Serie dont les données se repettent pour toutes les années. 
# C'est ce qui va définir le animation_group
fig = px.scatter(df, x = "gdpPercap", y = "lifeExp", animation_frame = "year", animation_group = "country",
                    size = "pop", color = "continent", hover_name = "country",
                    log_x = True, size_max = 55, range_x = [100,100000], range_y = [25,90])

fig.show()
```
