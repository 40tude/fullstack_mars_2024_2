## Challenge üí™

One of the challenge that Doctolib is facing is **last minute cancellations**. If a person is scheduling a doctor's appointment and actually ends not showing up, it's a waste of time (and money) for the doctor. Too many cancellations can lead to stop using Doctolib app. 

Therefore Doctolib would like to *build an algorithm that could predict whether a person is going to cancel his/her appointment* and reduce no-shows.

## Dataset 

Here is the dataset that you will be using for this exercise:

* [Doctolib Appointments Dataset](https://full-stack-assets.s3.eu-west-3.amazonaws.com/Deployment/doctolib_simplified_dataset_01.csv)

Here is an explanation of the columns:


- PatientId *Identification of a patient*
- AppointmentID: *Identification of each appointment*
- Gender: *Male or Female*
- AppointmentDay: *The day of the actuall appointment, when they have to visit the doctor.*
- ScheduledDay: *The day someone called or registered the appointment, this is before appointment of course.*
- Age: *How old is the patient.*
- Neighbourhood: *Where the appointment takes place.*
- Scholarship: *True of False*. This refers to the fact that the person has an insurance or not
- Hypertension: *True or False*
- Diabetes: *True or False*
- Alcoholism: *True or False*
- Handcap: *True or False*
- SMS_received: *1 or more messages sent to the patient.*
- No-show: *True or False.* **üéØ This is your target**
