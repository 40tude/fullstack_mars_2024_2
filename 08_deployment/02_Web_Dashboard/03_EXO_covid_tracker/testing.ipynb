```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
k_file_in = "./assets/data.csv"
k_nrows = 10_000

dataset = pd.read_csv(k_file_in, nrows=k_nrows)
dataset = dataset[dataset["cases"]>=0]

dataset["date"] = dataset["dateRep"].apply(lambda x: pd.to_datetime(x, format="%d/%m/%Y"))
dataset.drop(columns=["dateRep", "day", "month", "year"], inplace=True)
dataset.sort_values("date", ascending=True, inplace=True)

df  = dataset[["date", "cases"]].groupby("date").sum()
print(dataset.head())
```

```python
import datetime
from datetime import timedelta


df  = dataset[["date", "cases", "deaths", "countriesAndTerritories"]]
country = "Austria" #st.selectbox("Select a country", df["countriesAndTerritories"].sort_values().unique())
df = df[(df["countriesAndTerritories"]==country)]
df.sort_values("date", ascending=True, inplace=True)
df["sma_cases"] = df["cases"].rolling(7).mean()
df["sma_deaths"] = df["deaths"].rolling(7).mean()
df.set_index('date', inplace=True)

display(df.head(15))
print(df.iloc[0, 0])


date_of_interrest = datetime.date(2022, 1, 14)    #st.date_input("Select a date : ", )
date_of_interrest = pd.to_datetime(date_of_interrest, format="%Y-%m-%d")
print("Date of interrest : ", date_of_interrest.date())

date_before = date_of_interrest - timedelta(days=1)
print("The date before : ", date_before.date())

# avg_cases_today      = df[ (df["date"] == date_of_interrest.date()) ]["sma_cases"]
# avg_cases_today      = df[date_of_interrest.date()]["sma_cases"]
search_date = pd.to_datetime("2020-03-06", format="%Y-%m-%d")
print(search_date)
avg_cases_today      = df.loc[search_date,"sma_cases"]
print(avg_cases_today)
```
