# Frontend vs Backend 

## Context üìñ

[IBM](ibm.com) is one of the oldest tech companies in the world. Founded in 1911 in New York, they started building computers for the general public. They now serve over 170+ countries in the world. 

In a big company like this, one of the challenges is to attract and keep talents that will achieve great goals. Therefore, IBM conducted a research on their employees to study what were the different factors that would drive them to quit. 

## Goal üéØ

For this exercise, you are working within a team of developers & data scientists. Your management has a project for you: **build an interactive Dashboard** that will help HR know who are their employees and which ones are likely to quit. 

Your team is organized in the following way: 

1. A *Front-End Development Team* in charge of building the visible part of the Dashboard 
2. A *Back-End Development Team* in charge of getting data from different sources to serve the *Front-End Team*

You are part of the *Back-End Development Team*. Therefore **your goal is to build an API** that will manipulate IBM's data. 

After a quick meeting with the *Front-End Team*, it appears that they need the following endpoints:

* `/preview` - To be able to preview a few rows of the dataset
    * Front-End should be able to choose how many rows they want to preview 
    * There should be a default value 
* `/unique-values` - To be able to get unique values of a given column in the dataset

* `/groupby` - To be able to group by a given column and chose an aggregation metric 
    * Front-End should be able to choose which categorical column to use and which aggregation metric (`sum`, `mean`, `max`, `min`...)

* `/filter-by` - To be able to filter by one or several, categories within your dataset 
    * Front-End should be able to choose **one column** and **one or several categories** within that column 

* `/quantile` - To be able to retrieve the *top x%* or *bottom x%* values of a given column in the dataset 
    * *i.e* the top 10% `DailyRate` or the lowest 5% `DistanceFromHome`
    * Front-End should be able to choose the percentage and to choose whether it's the top or low values


## Constraints üöè

Your api must be documented so that your *Front-End Development Team* knows how to use it! A well documented API should at least contain:

* A global description of the app 
* A description of each endpoint
