# Object oriented (explicit) style 
* Sinon c'est le stype implicit/pyplot
 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.style.use("default")
```

# y est une liste

```python
y = [n**2 for n in range(0, 10)] 

fig, ax = plt.subplots()
ax.plot(y);                  # <- note le ; en fin de ligne 
```

```python
y = [n**2 for n in range(0, 10)] 

_, ax = plt.subplots()       # on se sert pas de fig on le remplace par _
ax.plot(y);                  
```

```python
_, ax = plt.subplots()
ax.plot([n**2 for n in range(0, 10)]);         # construit la liste lors de l'appel 
```

# x et y sont des listes 

```python
x = range(0, 10)
y = [n**2 for n in x] 

fig, ax = plt.subplots()
ax.plot(x, y);
```

```python
x = range(0, 10)
y = [n**2 for n in x] 

fig, ax = plt.subplots()
ax.scatter(x, y);
```

# y est un ndarray

```python
x = np.array(range(0,10))
y = x**2

fig, ax = plt.subplots()
ax.plot(y);
```

# x et y sont des ndarrays

```python
x = np.array(range(0,10))
y = x**2

fig, ax = plt.subplots()
ax.scatter(x, y);
```

```python
# on ne peut PAS passer à ax.plot un vecteur ligne
# on peut passer un vecteur colonne ou 
# si besoin .reshape(-1, 1) pour transformer une vecteur ligne en vecteur colonne
x = np.array(range(0,10))
y = x**2
print(f"dim : {y.ndim} shape: {y.shape}")
y = y.reshape(1, -1)
print(f"dim : {y.ndim} shape: {y.shape}")
y = y.reshape(-1, 1)
print(f"dim : {y.ndim} shape: {y.shape}")

fig, ax = plt.subplots()
ax.plot(y);

```

# y est un ndarray à 2 colonnes


```python
x = np.linspace(0, 2*np.pi, 100)
y = np.column_stack((np.sin(x), np.cos(x)))

fig, ax = plt.subplots()
ax.plot(y)
```

# Avec un Dataframe

```python
# on construit x et y en dehors du dataframe
x = np.array(range(-10,10))
y = x**2

my_data= {
  "x" : x,
  "y" : y
}
df = pd.DataFrame(my_data)
fig, ax = plt.subplots()
ax.scatter(data=my_data, x ="x", y="y");
```

```python
# on construit x et y dans le dataframe
my_data= {
  "x" : (x:=np.arange(-5, 5, 0.25)),
  "y" : x**2                          # np.cosh(x)
}
df = pd.DataFrame(my_data)
fig, ax = plt.subplots()
ax.scatter(data=my_data, x ="x", y="y");
```

```python
my_data= {
  "my_x" : (x := np.array(range(-10,10))),
  "my_y" : x**2
}

plt.style.use("seaborn-v0_8")
df = pd.DataFrame(my_data)
fig, ax = plt.subplots()
ax.scatter(data=my_data, x="my_x", y="my_y")
ax.set_xlabel(df["my_x"].name)
ax.set_ylabel(df["my_y"].name);
```
