# Bayes





* $ \text{P(B | A)} = \frac {\text{Surf Vert}}{\text{Surf Bleu}}$

* $ \text{P(B | A)} = \frac {  \text{P(A et B)}           }{\text{P(A)}}$    

* C'est la définition d'une proba conditionnelle

* $\text{P(A et B)} = \text{P(A)} * \text{P(B | A)}$
  * Pour avoir A et B, il faut déjà avoir A ET il faut avoir B (alors qu'on a déjà A) 



Par symétrie : 
* $\text{P(B et A)} = \text{P(B)} * \text{P(A | B)}$

D'où :
* $ \text{P(A | B)} = \frac{\text{P(B | A)} * \text{P(A)} } {\text{P(B)}}$

* $ \text{P(H | D)} = \frac{\text{P(D | H)} * \text{P(H)} } {\text{P(D)}}$


* Crédence = dégré de confiance qu'on accorde à une théorie (T ou H ci-dessus)

* P(H|D) : 
  * crédence a posteriori de la théorie sachant qu'on a observé les données

* P(D|H) : 
  * la vraissemblance des données
  * on suppose l'hypothèse ou la théorie vraie. 
  * Quelle est alors la probabilité d'observer les données?
  * la crédence qu'on accorde à ces données dans le cadre de cette théorie
  * peut valoir 1
  * faut imaginer un univers où la théorie est vraie et y chercher alors la proba d'y retrouver ces données
  * **IMPORTANT** 
    * la vraissemblance des données **N'EST PAS** la crédence de la théorie

* P(H)
  * la crédence à priori de la théorie/hypothèse
  * **prejugé** = jugement avant l'observation
  * a priori
  * "en l'état actuel de nos connaissances"

* P(D)
  * la probabilité des données
  * n'est généralement pas donnée
  * Penser à la formule des probabilités totales
  * $P(D) = \sum_{i \in I} P(D \cap A_i) = \sum_{i \in I} P(D | A_i) P(A_i) $ où Ai est une théorie/hypothèse alternative
  * Pour savoir si une théorie mérite notre confiance P(H|D), il faut la comparer à toutes les autres  
  * C'est l'objet du dénominateur

* Quelle est la probabilité qu'il a le cancer sachant qu'il a été diagnostiqué positif ?

  * $ P(\text{Il a le cancer} | \text{ Il est diagnostiqué avec le cancer})  = \text{P(H | D)} = \frac{\text{P(D | H)} * \text{P(H)} } {\text{P(D)}} $
  * On va quantifier nos **Prévisions** par rapport à la **Réalité**

### Avec la matrice 
* Il est diagnostiqué avec le cancer => on ne regarde que la seconde colonne
* Il a le cancer => on compare la cellule du bas à la somme de la colonne
* Précision $=\frac{TP}{TP + FP} = \frac{19}{19+3}$

### Avec Bayes
* $ \text{P(H | D)} = \frac{\text{P(D | H)} * \text{P(H)} } {\text{P(D)}}$
* $ P(\text{Il a le cancer} | \text{ Il est diagnostiqué avec le cancer}) = \frac{ P(\text{ Il est diagnostiqué avec le cancer} | \text{Il a le cancer} )  * P(\text{Il a le cancer})}{P(\text{Il est diagnostiqué avec le cancer})}$

* $ P(\text{ Il est diagnostiqué avec le cancer} | \text{Il a le cancer} ) = \frac{TP}{TP + FN}$
  * Il a le cancer => on ne regarde que la seconde ligne
  * En fait, on ici calcule le Recall

* $ P(\text{Il a le cancer}) = \frac{TP + FN}{\sum_T}$
  * C'est la somme de la seconde ligne sur toutes les cases

* $ P(\text{Il est diagnostiqué avec le cancer}) = \frac{TP + FP}{\sum_T}$
  * C'est la somme de la colonne de droite sur toutes les cases

* $ P(\text{Il a le cancer} | \text{ Il est diagnostiqué avec le cancer}) = \frac{\frac{TP}{TP + FN} * \frac{TP + FN}{\sum_T}}{\frac{TP + FP}{\sum_T}} = \frac{TP}{TP + FP}$
* On retrouve bien la Précision
