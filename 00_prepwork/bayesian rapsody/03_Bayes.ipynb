* $\text{P(B et A)} = \text{P(B)} * \text{P(A | B)}$

* $ \text{P(A | B)} = \frac{\text{P(B | A)} * \text{P(A)} } {\text{P(B)}}$

* $ \text{P(H | D)} = \frac{\text{P(D | H)} * \text{P(H)} } {\text{P(D)}}$

  * **IMPORTANT**

* Quelle est la probabilit qu'il a le cancer sachant qu'il a t diagnostiqu positif ?

  * $ P(\text{Il a le cancer} | \text{ Il est diagnostiqu avec le cancer})  = \text{P(H | D)} = \frac{\text{P(D | H)} * \text{P(H)} } {\text{P(D)}} $
  * On va quantifier nos **Prvisions** par rapport  la **Ralit**

### Avec la matrice 
* Il est diagnostiqu avec le cancer => on ne regarde que la seconde colonne
* Il a le cancer => on compare la cellule du bas  la somme de la colonne
* Prcision $=\frac{TP}{TP + FP} = \frac{19}{19+3}$

### Avec Bayes
* $ \text{P(H | D)} = \frac{\text{P(D | H)} * \text{P(H)} } {\text{P(D)}}$
* $ P(\text{Il a le cancer} | \text{ Il est diagnostiqu avec le cancer}) = \frac{ P(\text{ Il est diagnostiqu avec le cancer} | \text{Il a le cancer} )  * P(\text{Il a le cancer})}{P(\text{Il est diagnostiqu avec le cancer})}$

* $ P(\text{ Il est diagnostiqu avec le cancer} | \text{Il a le cancer} ) = \frac{TP}{TP + FN}$
  * Il a le cancer => on ne regarde que la seconde ligne
  * En fait, on ici calcule le Recall

* $ P(\text{Il a le cancer}) = \frac{TP + FN}{\sum_T}$
  * C'est la somme de la seconde ligne sur toutes les cases

* $ P(\text{Il est diagnostiqu avec le cancer}) = \frac{TP + FP}{\sum_T}$
  * C'est la somme de la colonne de droite sur toutes les cases

* $ P(\text{Il a le cancer} | \text{ Il est diagnostiqu avec le cancer}) = \frac{\frac{TP}{TP + FN} * \frac{TP + FN}{\sum_T}}{\frac{TP + FP}{\sum_T}} = \frac{TP}{TP + FP}$
* On retrouve bien la Prcision
